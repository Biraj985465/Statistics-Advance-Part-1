3. Write a Python function to calculate the probability distribution function (PDF) of a Bernoulli distribution.

```python
def bernoulli_pdf(x, p):
    """
    Calculate the PDF of a Bernoulli distribution.

    Parameters:
    - x : int (0 or 1) - the outcome
    - p : float (0 <= p <= 1) - probability of success

    Returns:
    - probability of outcome x
    """
    if x not in [0, 1]:
        raise ValueError("x must be 0 or 1 for Bernoulli distribution.")
    if not (0 <= p <= 1):
        raise ValueError("Probability p must be between 0 and 1.")

    return p**x * (1 - p)**(1 - x)

p_success = 0.7
print("P(X=1):", bernoulli_pdf(1, p_success))
print("P(X=0):", bernoulli_pdf(0, p_success))
```

9. Write a Python function to calculate Z-scores from a dataset and plot them.

```python
import numpy as np
import matplotlib.pyplot as plt

def calculate_and_plot_z_scores(data):
    """
    Calculates Z-scores of the input data and plots them.

    Parameters:
    - data: list or numpy array of numerical values

    Returns:
    - z_scores: numpy array of Z-scores
    """
    data = np.array(data)
    mean = np.mean(data)
    std_dev = np.std(data, ddof=0)

    z_scores = (data - mean) / std_dev

    plt.scatter(range(len(data)), z_scores, color='blue')
    plt.axhline(0, color='red', linestyle='--')
    plt.title('Z-scores of Dataset')
    plt.xlabel('Index')
    plt.ylabel('Z-score')
    plt.grid(True)
    plt.show()

    return z_scores

sample_data = [10, 12, 23, 23, 16, 23, 21, 16]
z = calculate_and_plot_z_scores(sample_data)
print("Z-scores:", z)
```

14. Write a Python program to calculate the Z-score for a given data point and compare it to a standard normal
distribution.

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm

def calculate_z_score(data, x):
    """
    Calculate the Z-score for a data point x given the data array.
    """
    mean = np.mean(data)
    std_dev = np.std(data, ddof=0)
    z = (x - mean) / std_dev
    return z

def plot_standard_normal_with_z(z):
    """
    Plot the standard normal distribution and mark the Z-score.
    """
    x = np.linspace(-4, 4, 1000)
    pdf = norm.pdf(x)

    plt.plot(x, pdf, label='Standard Normal Distribution')
    plt.fill_between(x, 0, pdf, where=(x <= z), color='skyblue', alpha=0.5, label=f'P(Z ≤ {z:.2f})')
    plt.axvline(z, color='red', linestyle='--', label=f'Z = {z:.2f}')

    plt.title('Standard Normal Distribution with Z-score')
    plt.xlabel('Z-score')
    plt.ylabel('Probability Density')
    plt.legend()
    plt.grid(True)
    plt.show()

data = [12, 15, 14, 10, 13, 15, 16, 11, 14, 13]
x = 18

z_score = calculate_z_score(data, x)
print(f"Z-score of data point {x}: {z_score:.3f}")

plot_standard_normal_with_z(z_score)
```

15.  Implement hypothesis testing using Z-statistics for a sample dataset.

```python
import numpy as np
from scipy.stats import norm

def z_test(sample_data, population_mean, population_std, alpha=0.05):
    """
    Perform a two-tailed Z-test for the mean.

    Parameters:
    - sample_data: list or array of sample observations
    - population_mean: hypothesized population mean (H0)
    - population_std: known population standard deviation
    - alpha: significance level (default 0.05)

    Returns:
    - z_stat: computed Z statistic
    - p_value: two-tailed p-value
    - conclusion: string stating whether to reject or fail to reject H0
    """
    n = len(sample_data)
    sample_mean = np.mean(sample_data)

    se = population_std / np.sqrt(n)

    z_stat = (sample_mean - population_mean) / se

    p_value = 2 * (1 - norm.cdf(abs(z_stat)))

    if p_value < alpha:
        conclusion = f"Reject the null hypothesis at α={alpha}."
    else:
        conclusion = f"Fail to reject the null hypothesis at α={alpha}."

    return z_stat, p_value, conclusion

sample = [12, 15, 14, 10, 13, 15, 16, 11, 14, 13]
pop_mean = 13
pop_std = 2
alpha_level = 0.05

z_statistic, p_val, result = z_test(sample, pop_mean, pop_std, alpha_level)

print(f"Z-statistic: {z_statistic:.3f}")
print(f"P-value: {p_val:.4f}")
print(result)
```

16. Create a confidence interval for a dataset using Python and interpret the result.

```python
import numpy as np
from scipy import stats

def confidence_interval(data, confidence=0.95):
    """
    Calculate the confidence interval for the mean of the data.

    Parameters:
    - data: list or numpy array of sample observations
    - confidence: confidence level (default 0.95 for 95%)

    Returns:
    - (lower_bound, upper_bound): tuple of confidence interval bounds
    """
    data = np.array(data)
    n = len(data)
    mean = np.mean(data)
    sem = stats.sem(data)
    t_crit = stats.t.ppf((1 + confidence) / 2, df=n-1)

    margin_of_error = t_crit * sem
    lower_bound = mean - margin_of_error
    upper_bound = mean + margin_of_error

    return lower_bound, upper_bound

sample_data = [12, 15, 14, 10, 13, 15, 16, 11, 14, 13]
conf_level = 0.95

lower, upper = confidence_interval(sample_data, confidence=conf_level)
mean_val = np.mean(sample_data)

print(f"Sample Mean = {mean_val:.2f}")
print(f"{int(conf_level*100)}% Confidence Interval: ({lower:.2f}, {upper:.2f})")
print("Interpretation: We are 95% confident that the true population mean lies within this interval.")
```

17. Generate data from a normal distribution, then calculate and interpret the confidence interval for its mean.

```python
import numpy as np
from scipy import stats

np.random.seed(42)
mu, sigma = 50, 5
sample_size = 30
data = np.random.normal(mu, sigma, sample_size)

def confidence_interval(data, confidence=0.95):
    n = len(data)
    mean = np.mean(data)
    sem = stats.sem(data)
    t_crit = stats.t.ppf((1 + confidence) / 2, df=n-1)
    margin = t_crit * sem
    return mean - margin, mean + margin

lower, upper = confidence_interval(data, confidence=0.95)
mean_val = np.mean(data)

print(f"Sample mean: {mean_val:.2f}")
print(f"95% Confidence Interval for the mean: ({lower:.2f}, {upper:.2f})")
print("Interpretation: We are 95% confident that the true population mean lies within this interval.")
```

24. Write a Python function to calculate the skewness and kurtosis of a dataset generated from a normal
distribution.

```python
import numpy as np
from scipy.stats import skew, kurtosis

def calculate_skewness_kurtosis(mean=0, std=1, size=1000):
    """
    Generates data from a normal distribution and calculates skewness and kurtosis.

    Parameters:
    - mean: mean of the normal distribution
    - std: standard deviation of the normal distribution
    - size: number of data points

    Returns:
    - skewness: measure of asymmetry
    - kurt: measure of tail heaviness (excess kurtosis)
    """
    data = np.random.normal(loc=mean, scale=std, size=size)
    skewness = skew(data)
    kurt = kurtosis(data)

    return skewness, kurt

skewness_val, kurtosis_val = calculate_skewness_kurtosis(mean=0, std=1, size=1000)
print(f"Skewness: {skewness_val:.4f}")
print(f"Kurtosis (excess): {kurtosis_val:.4f}")
```
